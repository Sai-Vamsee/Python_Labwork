##                                      Visualizing the Distributions

```python
import numpy as np
import matplotlib.pyplot as plt
from math import *
```

### Binomial Distribution

```python
def binomial(n,p,x):
    return comb(n,x)*p**x*(1-p)**(n-x)

n = int(input("Enter the x limit:"))
p = float(input("Enter Probability :"))

X = [i for i in range(n+1)]
y = []
for x in X:
    y.append(binomial(n,p,x))
    
plt.title('Plot of Binomial Random Variable')
plt.xlabel('x', fontsize = 15)
plt.ylabel('f(x)', fontsize = 15)   
plt.plot(X,y)
plt.show()
```

### Normal Distribution

```python
def normal(a,b):
    
    x= np.linspace(a-4*b,a+4*b,100)
    f= lambda x: (1/b*sqrt(2*pi))*e**((-1/2*(((x-a)/b)**2)))
    plt.figure(figsize = (10,5))
    plt.scatter(a,f(a))
    plt.bar(x,f(x))
    plt.xlabel("x",fontsize = 15)
    plt.ylabel("f(x)",fontsize = 15)
    plt.title("Plot of Normal Distribution ")

a = int(input("Enter the Mean Value : "))
b = int(input("Enter the Variance Value : "))
normal(a,b)
```

### Poisson Distribution

```python
def poisson(lam):
    
    x= np.arange(100)
    f= lambda x: np.exp(-lam)*(lam**x / factorial(x))
    y = []
    for i in range (len(x)):
        y.append(f(x[i]))
    plt.plot(x,y)
    plt.xlabel("x",fontsize = 15)
    plt.ylabel("f(x)",fontsize = 15)
    plt.title("Plot of Poisson Distribution ")

lam = float(input("Enter the Lambda Value : "))

poisson(lam)
```
